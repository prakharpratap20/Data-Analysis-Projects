## String Handling

```python
oo[oo.Athlete.str.contains("Florence")]
```

#### In which events did Jesse Owens win a medal?

```python
jo = oo[oo.Athlete == "OWENS, Jesse"]
```

```python
jo.Event.value_counts()
```

#### Which country has won the most men's Gold medals in single badminton over the year?
##### Sort the results alphabetically by the players names

```python
gbm = oo[(oo.Medal == "Gold") & (oo.Gender == "Men") & (oo.Sport == "Badminton")]
```

```python
gbm.sort_values(by=["NOC", "Athlete"])
```

### Which three countries have won the most medals in recent years(from 1984 to 2008)?

```python
rec = oo[oo.Edition >=  1984]
rec.NOC.value_counts().head(3)
```

### Display the male gold medal winners for the 100m Track & Field sprint event over the years. List the results starting with the most recent. Show the Olympic City, Edition, Athlete and the country they represent

```python
gmh = oo[(oo.Gender == "Men") & (oo.Medal == "Gold") & (oo.Event == "100m")]
gmh.sort_values("Edition", ascending=False)[["City", "Edition", "Athlete", "NOC"]]
```

## Plotting

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

What were the different sporst in the first olympics? Plot

```python
fo = oo[oo.Edition == 1896]
fo.head()
```

Line Plot

```python
fo.Sport.value_counts().plot(kind="line", figsize=(10, 3));
```

Bar Plot

```python
fo.Sport.value_counts().plot(kind="bar");
```

Horizontal bar plot

```python
fo.Sport.value_counts().plot(kind="barh", color="plum");
```

Pie Chart

```python
fo.Sport.value_counts().plot(kind="pie", colormap="Pastel1");
```
