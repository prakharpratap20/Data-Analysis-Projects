```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt  
import seaborn as sns
```

## Performing Sentiment Analysis

```python
from textblob import TextBlob
```

```python
comments.head(6)
```

```python
polarity = []

for comment in comments["comment_text"]:
    try:
        polarity.append(TextBlob(comment).sentiment.polarity)
    except:
        polarity.append(0)
```

```python
len(polarity)
```

```python
comments["polarity"] = polarity
```

```python
comments.head(5)
```

## Word Cloud Analysis of Data

```python
filter1 = comments["polarity"] == 1
```

```python
comments_positive = comments[filter1]
```

```python
filter2 = comments["polarity"] == -1
```

```python
comments_negative = comments[filter2]
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
set(STOPWORDS)
```

```python
comments["comment_text"]
```

```python
type(comments["comment_text"])
```

```python
total_comments_positive = " ".join(comments_positive["comment_text"])
```

```python
wordcloud = WordCloud(stopwords=set(STOPWORDS)).generate(total_comments_positive)
```

```python
plt.imshow(wordcloud)
plt.axis("off")
```

```python
total_comments_negative = " ".join(comments_negative["comment_text"])
```

```python
wordcloud2 = WordCloud(stopwords=set(STOPWORDS)).generate(total_comments_negative)
```

```python
plt.imshow(wordcloud2)
plt.axis("off")
```

## Performing Emoji Analysis

```python
import emoji
```

```python
emoji_list = []

for comment in comments["comment_text"]:
    for char in comment:
        if char in emoji.EMOJI_DATA:
            emoji_list.append(char)
```

```python
from collections import Counter
```

```python
Counter(emoji_list)
```

```python
emojis = [Counter(emoji_list).most_common(10)[i][0] for i in range(10)]
```

```python
freqs = [Counter(emoji_list).most_common(10)[i][1] for i in range(10)]
```

```python
plt.bar(emojis, freqs)
plt.xlabel('Emojis')
plt.ylabel('Frequencies')
plt.title('Top 10 Emojis Frequency')
```

## Collect Entire data of Youtube

```python
import os
```

```python
files = os.listdir(r"additional_data")
```

```python
files
```

```python
files_csv = [file for file in files if ".csv" in file]
```

```python
files_csv
```

```python
import warnings
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
full_df = pd.DataFrame()
path = "additional_data"

for file in files_csv:
    current_df = pd.read_csv(path+"/"+file, encoding="iso-8859-1", on_bad_lines="skip")
    full_df = pd.concat([full_df, current_df], ignore_index=True)
```

```python
full_df.shape
```

## export data into CSV

```python
full_df[full_df.duplicated()].shape
```

```python
full_df = full_df.drop_duplicates()
```

```python
full_df.shape
```

```python
full_df.to_csv(r"youtube_data_csv.csv", index=False)
```

```python
full_df.to_json(r"youtube_data_json.json", index=False)
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine(r"sqlite:///youtube_sqilite.sqlite")
```

```python
full_df.to_sql("Users", con=engine, if_exists="append")
```

## which category has the max likes

```python
full_df.head(5)
```

```python
json_df = pd.read_json(r"additional_data/US_category_id.json")
```

```python
json_df
```

```python
json_df["items"][0]
```

```python
cat_dict = {}

for item in json_df["items"].values:
    cat_dict[int(item["id"])] = item["snippet"]["title"] 
```

```python
cat_dict
```

```python
full_df["category_name"] = full_df["category_id"].map(cat_dict)
```
